```python
# Your answer here

#1. **List Your Favorite Artists**:
#    - Make a list of your three favorite artists and store it in a variable named `artists`.

artists = ["Billy Raffoul", "Craig David", "Gerilson Insrael"]
```

```python
# Function to get the top tracks of an artist
SPOTIFY_CLIENT_ID = "4109b2d9f5014671863bb2df1d1fbdd3"
SPOTIFY_CLIENT_SECRET = "5cb719fae80c4191a9c0b288f51bfe50"

def get_spotify_token(client_id, client_secret):
    auth_url = 'https://accounts.spotify.com/api/token'
    auth_response = requests.post(auth_url, {
        'grant_type': 'client_credentials',
        'client_id': client_id,
        'client_secret': client_secret,
    })
    auth_response_data = auth_response.json()
    return auth_response_data['access_token']

def get_top_tracks(artist_name):
    token = get_spotify_token(SPOTIFY_CLIENT_ID, SPOTIFY_CLIENT_SECRET)
    headers = {
        'Authorization': f'Bearer {token}'
    }
    search_url = 'https://api.spotify.com/v1/search'
    params = {
        'q': artist_name,
        'type': 'artist',
        'limit': 1
    }
    response = requests.get(search_url, headers=headers, params=params)
    artist_data = response.json()
    if artist_data['artists']['items']:
        artist_id = artist_data['artists']['items'][0]['id']
        top_tracks_url = f'https://api.spotify.com/v1/artists/{artist_id}/top-tracks'
        params = {
            'market': 'US'
        }
        response = requests.get(top_tracks_url, headers=headers, params=params)
        top_tracks_data = response.json()
        top_tracks = [track['name'] for track in top_tracks_data['tracks'][:5]]
        return top_tracks
    else:
        return []
    
    # Get top tracks for each favorite artist
top_tracks_list = []
for artist in artists:
    top_tracks = get_top_tracks(artist)
    top_tracks_list.append(top_tracks)

top_tracks_list
```

```python
#function to find related artists.
def find_related_artists(artist_name):
    token = get_spotify_token(SPOTIFY_CLIENT_ID, SPOTIFY_CLIENT_SECRET)
    headers = {
        'Authorization': f'Bearer {token}'
    }
    search_url = 'https://api.spotify.com/v1/search'
    params = {
        'q': artist_name,
        'type': 'artist',
        'limit': 1
    }
    response = requests.get(search_url, headers=headers, params=params)
    artist_data = response.json()
    if artist_data['artists']['items']:
        artist_id = artist_data['artists']['items'][0]['id']
        related_artists_url = f'https://api.spotify.com/v1/artists/{artist_id}/related-artists'
        response = requests.get(related_artists_url, headers=headers)
        related_artists_data = response.json()
        related_artists = [artist['name'] for artist in related_artists_data['artists'][:5]]
        return related_artists
    else:
        return []
```

```python
# Find related artists for each favorite artist
related_artists_list = []
for artist in artists:
    related_artists = find_related_artists(artist)
    related_artists_list.append(related_artists)

related_artists_list
```

```python
# List to store the combined top tracks
combined_playlist = []

# Add top tracks of favorite artists to the playlist
for tracks in top_tracks_list:
    combined_playlist.extend(tracks)

# Add top tracks of related artists to the playlist
for related_artists in related_artists_list:
    for related_artist in related_artists:
        related_artist_top_tracks = get_top_tracks(related_artist)
        combined_playlist.extend(related_artist_top_tracks)

combined_playlist
```

### Getting Artists from a Playlist

To extract all the artists from the tracks in a playlist, you'd typically follow these steps:

1. Fetch the playlist's tracks.
2. Iterate through each track.
3. For each track, extract the associated artists.

```python
# List to store unique artist names
artists_list = []

for track_item in tracks['items']:
    track = track_item['track']
    for artist in track['artists']:
        artist_name = artist['name']
        if artist_name not in artists_list:  # This ensures each artist is added only once
            artists_list.append(artist_name)

print(artists_list)
```

4. **Artistic Flair**:
   - Create a dictionary where the keys are the names of the first 10 tracks, and the values are lists containing the names of the artists associated with each track.
   - For example: `{"TrackName1": ["Artist1", "Artist2"], "TrackName2": ["Artist3"]}`
   

```python
# Your answer here
def get_spotify_token(client_id, client_secret):
    auth_url = 'https://accounts.spotify.com/api/token'
    auth_response = requests.post(auth_url, {
        'grant_type': 'client_credentials',
        'client_id': client_id,
        'client_secret': client_secret,
    })
    auth_response_data = auth_response.json()
    return auth_response_data['access_token']

# Fetch the token
token = get_spotify_token(SPOTIFY_CLIENT_ID, SPOTIFY_CLIENT_SECRET)

# Set up the headers for API requests
headers = {
    'Authorization': f'Bearer {token}'
}

# 1. Featured Exploration
def fetch_featured_playlists():
    featured_playlists_url = 'https://api.spotify.com/v1/browse/featured-playlists'
    response = requests.get(featured_playlists_url, headers=headers)
    playlists_data = response.json()
    return playlists_data['playlists']['items']

def display_top_playlists(playlists, top_n=5):
    top_playlists = playlists[:top_n]
    for idx, playlist in enumerate(top_playlists):
        print(f"{idx + 1}. {playlist['name']} (ID: {playlist['id']})")
    return top_playlists

featured_playlists = fetch_featured_playlists()
top_playlists = display_top_playlists(featured_playlists)
```

```python
# 2. Deep Dive
# Choose one playlist (you can choose based on your preference)
chosen_playlist = top_playlists[4]  # Replace with your choice if needed

def fetch_playlist_details(playlist_id):
    playlist_url = f'https://api.spotify.com/v1/playlists/{playlist_id}'
    response = requests.get(playlist_url, headers=headers)
    return response.json()

playlist_details = fetch_playlist_details(chosen_playlist['id'])
print(f"Name: {playlist_details['name']}")
print(f"Description: {playlist_details['description']}")
print(f"Total Tracks: {playlist_details['tracks']['total']}")
```

```python
# 3. Track-tastic
def fetch_playlist_tracks(playlist_id, limit=10):
    tracks_url = f'https://api.spotify.com/v1/playlists/{playlist_id}/tracks'
    params = {
        'limit': limit
    }
    response = requests.get(tracks_url, headers=headers, params=params)
    tracks_data = response.json()
    return tracks_data['items']

playlist_tracks = fetch_playlist_tracks(chosen_playlist['id'])
track_names = [track['track']['name'] for track in playlist_tracks]
print(f"First 10 Tracks: {track_names}")
```

```python
# 4. Artistic Flair
def create_track_artist_dict(tracks):
    track_artist_dict = {}
    for track in tracks:
        track_name = track['track']['name']
        artist_names = [artist['name'] for artist in track['track']['artists']]
        track_artist_dict[track_name] = artist_names
    return track_artist_dict

track_artist_dict = create_track_artist_dict(playlist_tracks)
print(f"Track-Artist Dictionary: {track_artist_dict}")
```
